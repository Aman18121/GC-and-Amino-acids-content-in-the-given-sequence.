```python
'''blast - BLAST (Basic Local Alignment Search Tool) is a bioinformatics algorithm used to compare biological sequences such as DNA, RNA, or 
protein against databases to identify regions of similarity. It provides a method to search for evolutionary relationships, annotate genes and proteins, 
and diagnose genetic disorders by aligning and scoring sequence matches based on statistical significance. BLAST is pivotal in deciphering biological 
functions, understanding evolutionary processes, and advancing biomedical research through efficient sequence analysis.'''
```

```python
# This a basic python function which acts like blast and finds out the matches between two sequences, hence its not linked with  any databases so
# it cant provide the same data as NCBI Blast and learning heuristic methods
# This was made by my during the start of my Life science informatic course 

def matches():
    number_matches = 0
    positions =[]
    s1 = str(input("Enter strand"))
    s2 = str(input("Enter strand query"))
    if len(s1) != len(s2):
        print("The strands are of different lengths.")
        return None
    for i in range(len(s1)):
        if s1[i] == s2[i]:
            number_matches += 1
            positions.append(i)
     #percentage of match.
    percentage =(number_matches/ len(s1) * 100) 

    print("percentage of match   {}".format(percentage))
    print(positions)        
    return number_matches       
matches()
            
```

```python
'''GC content, the percentage of guanine (G) and cytosine (C) bases in a DNA or RNA sequence, is important because it affects the stability of the 
molecule. Sequences with higher GC content typically have higher melting temperatures, making them more stable under varying environmental conditions.
This stability is crucial for processes like DNA replication and PCR (Polymerase Chain Reaction). Additionally, GC content can indicate evolutionary
relationships between species, as organisms with similar GC content may share common ancestry or environmental adaptations. In biotechnology, GC 
content influences primer design and probe selection for accurate and efficient DNA amplification and detection techniques.'''
```

```python
#This code identifies the conc of G and C nucleic acid in the given sequence 
# Nothing is better than interpreting data in a well-visualized form.




def G_C_counter (query_seq1):
    main_seq1 = list(query_seq1)
    main_seq2 = {}
    for item in main_seq1:
        if item == "" or item == ' ' or item == '' or item == " ":
            main_seq1.remove(item)
        if item in main_seq2:
            main_seq2[item] += 1
        else :
             main_seq2[item] = 1

    print(main_seq2)
   # print(main_seq1)
    GC_amino1 =  ((main_seq1.count("G") +  main_seq1.count("C"))/ len(main_seq1) * 100)
    print ("The percentage of GC content in the sequence is {}" .format(GC_amino1))

    plt.figure(figsize=(6, 6))
    plt.bar(list(main_seq2.keys()) , list(main_seq2.values()), color="red", alpha=0.5)
    plt.grid(True)
    plt.ylabel("number of nucleotides present")
    plt.xlabel("nucleotides (AGCT)")
    plt.yticks( rotation = 45)
    plt.title("ATGC content in the given seqeunce")
    plt.show

G_C_counter("GCATTCTGAGGCATTCTCTAACAGGTCCTCACCGTGCTGGCCCTGCTGGCCCTCTGGGGACCCAACTCTGTTCAGGCCTATTCCAGCCAGCACCTGTGCGGCTCCAACCTAGTGGAGGCACTGTACATGACATGTGGACGGAGTGGCTTCTATAGACCCCACGACCGCCGAGAGCTGGAGGACCTCCAGGTGGAGCAGGCAGAACTGGGTCTGGAGGCAGGCGGCCTGCAGCCTTCGGCCCTGGAGATGATTCTGCAGAAGCGCGGCATTGTGGATCAGTGCTGTAATAACATTTGCACATTTAACCAGCTGCAGAACTACTGCAATGTCCCTTAGACACCTGCCTTGGGCCTGGCCTGCTGCTCTGCCCTGGCAACCAATAAACCCCTTGAATGAG")
            
    
```
